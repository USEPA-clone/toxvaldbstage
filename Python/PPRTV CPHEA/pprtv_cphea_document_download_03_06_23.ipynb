```python
import requests
import pandas as pd
import time as t
import urllib.request
import os.path
from PIL import Image
from io import BytesIO
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By


file = r"C:\Users\mmille16\OneDrive - Environmental Protection Agency (EPA)\Profile\Downloads\pprtv_cphea_chemicals_20230123.xlsx"
cphea = pd.read_excel(file)
```

```python
#Hold chemical names and urls for chemicals whose pdf url was invalid
retry = {}

#Save location of origin pdfs
save_path = r"L:\PRIV\ToxValDB\Document Webscrape\pprtv_cphea\PPRTV CPHEA\Origin Documents"

for i in range(len(cphea)):
    chem_name = cphea.loc[i,'Chemical']
    file_name = f'{save_path}/{chem_name}_pprtv_cphea_origin.pdf' #Save name of support pdf
    if os.path.isfile(file_name): #In case this needs to be run again, make sure I don't double download a file
        continue
    #Extract the pdf url from the excel file
    url = cphea.loc[i,'PPRTV Assessment']
    try:
        t.sleep(1)
    #Access the pdf url, open a file in the pprtv cphea pdf folder, write, and save the file from the pdf url
        response = urllib.request.urlopen(url)
    except Exception as e:
        retry[cphea.loc[i,'PPRTV Substance ID']] = chem_name
    #In event of possible errors in url, let me know please and thank you.
        #print(f'{chem_name} url could not be loaded. Error response is: {e} \n url is {url}')
        continue
    file = open(file_name,'wb')
    file.write(response.read())
    file.close()
    

#Might be able to find more valid pprtv pdf links on the pprtv_chpea webpage
for sub_id in retry:
    file_name = f'{save_path}/{retry[sub_id]}_pprtv_cphea.pdf'
    if os.path.isfile(file_name): #In case this needs to be run again, make sure I don't double download a file
        print(f'the pdf for {chem_name} is already saved in the directory.')
        continue
#Slenium browser settings
    browser = f'https://cfpub.epa.gov/ncea/pprtv/chemicalLanding.cfm?pprtv_sub_id={sub_id}'
    Options = webdriver.ChromeOptions()
    Options.headless = True
    s=Service(r'C:\Users\mmille16\Documents\chromedriver.exe')
    driver = webdriver.Chrome(service=s, options = Options)
    driver.get(browser)
    
#Look for the webpage element with the pprtv link
    try:
        item = driver.find_element(By.PARTIAL_LINK_TEXT,'Toxicity')
    except:
#Some may not have one. Tell me just in case
        print(f'{retry[sub_id]} webpage does not have a link to the pprtv pdf')
        continue
#Travel through the webpage until we arrive at the downloadable pdf link
    item.click()
    t.sleep(0.5)
    tab = driver.find_element(By.PARTIAL_LINK_TEXT,'Downloads')
    tab.click()
    t.sleep(0.5)
#Get the pdf url and save as a pdf in the directory
    doc = driver.find_element(By.PARTIAL_LINK_TEXT,'Toxicity')
    response = urllib.request.urlopen(doc.get_attribute('href'))
    file = open(file_name, 'wb')
    file.write(response.read())
    file.close()
    driver.close()
print('done')
```

```python
#Puts both of the methods in the cell above into a script with only one for loop.

cphea_file = r"C:\Users\mmille16\OneDrive - Environmental Protection Agency (EPA)\Profile\Downloads\pprtv_cphea_chemicals_20230123.xlsx"
cphea = pd.read_excel(cphea_file)
retry = {}
save_path = r"C:\Users\mmille16\OneDrive - Environmental Protection Agency (EPA)\Profile\Desktop\pprtv_cphea_pdfs" #Location of PPRTV CPHEA pdf folder

for i in range(len(cphea)):
    chem_name = cphea.loc[i,'Chemical']
    file_name = f'{save_path}/{chem_name}_pprtv_cphea_origin.pdf' #Save name of support pdf
    if os.path.isfile(file_name): #In case this needs to be run again, make sure I don't double download a file
        #print(f'the pdf for {chem_name} is already saved in the directory.')
        continue
    #Extract the pdf url from the excel file
    url = cphea.loc[i,'PPRTV Assessment']
    try:
        t.sleep(1)
    #Access the pdf url, open a file in the pprtv cphea pdf folder, write, and save the file from the pdf url
        response = urllib.request.urlopen(url)
    except:
        sub_id = cphea.loc[i,'PPRTV Substance ID']
        browser = f'https://cfpub.epa.gov/ncea/pprtv/chemicalLanding.cfm?pprtv_sub_id={sub_id}'
        Options = webdriver.ChromeOptions()
        Options.headless = True
        s=Service(r'C:\Users\mmille16\Documents\chromedriver.exe')
        driver = webdriver.Chrome(service=s, options = Options)
        driver.get(browser)
        print(f'pdf link for {chem_name} was invalid. Checking webpage for updated pdf link.')
        try:
            item = driver.find_element(By.PARTIAL_LINK_TEXT,'Toxicity')
            item.click()
            t.sleep(0.5)
            tab = driver.find_element(By.PARTIAL_LINK_TEXT,'Downloads')
            tab.click()
            t.sleep(0.5)
            doc = driver.find_element(By.PARTIAL_LINK_TEXT,'Toxicity')
            response = urllib.request.urlopen(doc.get_attribute('href'))
            driver.close()
            print(f'pdf link for {chem_name} was located on the webpage.')
        except:
            driver.close()
            print(f'could not find a pprtv pdf for {chem_name}.')
            continue
    file = open(f'{save_path}/{chem_name}_pprtv_chpea_origin.pdf','wb')
    file.write(response.read())
    file.close()
```

```python
#PNG Screenshots Screenshots

#Selenium Browser settings
Options = webdriver.ChromeOptions()
Options.headless = True
s=Service(r'C:\Users\mmille16\Documents\chromedriver.exe')
driver = webdriver.Chrome(service=s, options = Options)

#Location of PPRTV CPHEA pdf folder
save_path = r"L:\PRIV\ToxValDB\Document Webscrape\pprtv_cphea\PPRTV CPHEA\Extraction Documents\Extraction PNG" 

#Loop through the ids, access the webpage, and take a screenshot of the information
for i in range(len(cphea)):
    chem_name = cphea.loc[i,'Chemical']
    file_name = f'{save_path}\\{chem_name}_pprtv_cphea_extraction.png'
    if os.path.isfile(file_name):#In case this needs to be run again, make sure I don't double download a file
        continue
    try:
#Open the webpage based on the cphea substance id
        sub_id = cphea.loc[i,'PPRTV Substance ID']
        browser = f'https://cfpub.epa.gov/ncea/pprtv/chemicalLanding.cfm?pprtv_sub_id={sub_id}'
        driver.get(browser)
#Scroll to the bottom of the webpage
        S = lambda X: driver.execute_script('return document.body.parentNode.scroll'+X)
        t.sleep(1)
#Main content element of the webpage
        element2 = driver.find_element(By.ID, "main-content")
#Set the window size to see and  screenshot the entire main content element
        driver.set_window_size(800,S("Height"))
        element2.screenshot(file_name)
    except Exception as e: #In case a webpage cannot be accessed, let me know please and thank you
        print(f'A problem arose for {chem_name} \n {browser} \n {e}')
        continue
driver.close()
print('done')
```

```python
#Convert png screenshots to pdfs
save_path = r"L:\PRIV\ToxValDB\Document Webscrape\pprtv_cphea\PPRTV CPHEA\Extraction Documents\Extraction PNG" #Location of PPRTV CPHEA png folder
out_path = r"L:\PRIV\ToxValDB\Document Webscrape\pprtv_cphea\PPRTV CPHEA\Extraction Documents\Extraction PDF" #Save location for pdf screenshots
for png in os.listdir(save_path):
#In case there are any files that are not pngs in the directory: ignore them
    if str(png).lower().find('.png') == -1:
        continue
#In case this needs to be run again, don't redo work I've already done.
    if os.path.isfile(os.path.join(out_path,png.replace('.png','.pdf'))):
        continue
#Open the png image and convert it to a pdf to save in the new directory
    im1 = Image.open(os.path.join(save_path,png))
    im_1 = im1.convert('RGB')
    im_1.save(os.path.join(out_path,png.replace('.png','.pdf')))
```
