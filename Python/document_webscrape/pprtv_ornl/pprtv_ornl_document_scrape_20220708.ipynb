```python
#Final script to take screenshots of the webpages for each chemical in hhpprtv.ornl.gov
import requests
import pandas as p
import numpy as np
import time as t
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from PIL import Image
from io import BytesIO
from selenium.webdriver.support.ui import Select

file_path = r'L:\PRIV\ToxValDB\Document Webscrape\pprtv_ornl\output\webpage_screenshots'


#This set of commands opens the browser url.
Options = webdriver.ChromeOptions()
Options.headless = True
s=Service(r'C:\Users\mmille16\Documents\chromedriver.exe')
driver = webdriver.Chrome(service=s, options = Options)
URL = 'https://hhpprtv.ornl.gov/quickview/pprtv.php'
driver.get(URL)


#This set of commands is to click on the drop down menu and select the options (manual selection for testing purposes)
select = Select(driver.find_element(By.ID, "mySelect"))
options = select.options
for i in range (1,len(options)):
    t.sleep(1)
    select = Select(driver.find_element(By.ID, "mySelect"))
    options = select.options
    name = options[i].text
    t.sleep(1)
    options[i].click()
    a = t.localtime()
    current_time = t.strftime("%Y%m%d_%H%M%S",a)
    t.sleep(1)
    S = lambda X: driver.execute_script('return document.body.parentNode.scroll'+X)
    driver.set_window_size(S('Width'),S('Height')) # May need manual adjustment                                                                                                                
    p = driver.find_element(By.TAG_NAME,'body').screenshot('pprtv_ornl_'+name+'_webpage_'+current_time+'.png')
    t.sleep(1)
    element2 = driver.find_element(By.CLASS_NAME, "field-container")
    location = element2.location
    size = element2.size
    right = location['x'] + size['width']
    left = location['x'] - 50 #slight manual adjustment to see the page more clearly
    top = location['y']
    bottom = location['y'] + size['height']

#image cropping of screenshot
    im = Image.open('pprtv_ornl_'+name+'_webpage_'+current_time+'.png')
    im1 = im.crop((left, top, right, bottom))
    im2 = im1.save(file_path+'\pprtv_ornl_'+name+'_webpage_'+current_time+'.png')

driver.quit()
    
```

```python
#Final script to download the support documents for each chemical
#Had to be run in batches of ~60-80 to avoid timeout errors for urllib requesting
#Exception for the chemical Benz[a]anthracene. This url did not work as written it appeared in the webpage. 
#Correction for this exception explained in the cell below

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import urllib.request

Options = webdriver.ChromeOptions()
Options.headless = True
s=Service(r'C:\Users\mmille16\Documents\chromedriver.exe')
driver = webdriver.Chrome(service=s, options = Options)
url='https://hhpprtv.ornl.gov/quickview/pprtv.php'
driver.get(url)
file_path = 'L:\PRIV\ToxValDB\Document Webscrape\pprtv_ornl\output\support_pdfs'


t.sleep(5)
select = Select(driver.find_element(By.ID, "mySelect"))
options = select.options

for i in range (1,len(options)):
    t.sleep(5)
    select = Select(driver.find_element(By.ID, "mySelect"))
    options = select.options
    name = options[i].text
    select.select_by_index(i)
    link = driver.find_element(By.PARTIAL_LINK_TEXT,'PPRTV')
    t.sleep(1)
    
    doc_url = link.get_attribute('href')
    a = t.localtime()
    current_time = t.strftime("%Y%m%d_%H%M%S",a)
    t.sleep(1)
    
    response = urllib.request.urlopen(doc_url)

    file = open(file_path+'\\pprtv_ornl_'+name+'_summary_'+current_time+'.pdf', 'wb')
    file.write(response.read())
    file.close()

driver.quit()
```

```python
#Benz[a]anthracene's link did not work properly, needed to remove the brackets from the [a] to make the url work

Options = webdriver.ChromeOptions()
Options.headless = True
s=Service(r'C:\Users\mmille16\Documents\chromedriver.exe')
driver = webdriver.Chrome(service=s, options = Options)
url='https://hhpprtv.ornl.gov/quickview/pprtv.php'
driver.get(url)
file_path = 'L:\PRIV\ToxValDB\Document Webscrape\pprtv_ornl\output\support_pdfs'


t.sleep(5)
select = Select(driver.find_element(By.ID, "mySelect"))
options = select.options

select.select_by_index(32)

link = driver.find_element(By.PARTIAL_LINK_TEXT,'PPRTV')
t.sleep(1)
    
doc_url = link.get_attribute('href')
doc_url = doc_url.replace("[","").replace("]","")

t.sleep(1)
    
response = urllib.request.urlopen(doc_url)

file = open(file_path+'\\pprtv_ornl_'+name+'_summary_'+current_time+'.pdf', 'wb')
file.write(response.read())
file.close()
```
